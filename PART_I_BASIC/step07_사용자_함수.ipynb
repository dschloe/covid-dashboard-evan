## Docstring

```python
def add(a, b):
    """a, b를 더하는 함수

    Parameters:
    a (int): int형 숫자 a가 입력
    
    b (int): int형 숫자 b가 입력

    Returns:
    int: 반환 값

   """   
    c = a + b
    return c

print(add.__doc__)
```
